```python
import pandas as pd

df = pd.read_csv("./data/SalmonPriceFRED.csv")
df.head()
```

```python
df.columns = ['date', 'price']
df = df.dropna()
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
# plot historic salmon prices
import matplotlib.pyplot as plt

plt.plot(df['date'], df['price'])
plt.xlabel('date')
plt.ylabel('price $')
plt.title('Historic Salmon Prices')
plt.grid(True)
plt.show()
```

```python
import re

sum([1 if re.search(r"2000", str(x)) else 0 for x in list(df['date'])])
```

```python
# is there just a lot of missing data
import numpy as np

log_returns = np.log(df['price'] / df['price'].shift(1))
print(f"Estimated Salmon Volitility {float(log_returns.std() * np.sqrt()) * 100:.2f}%")
```
